## Dataset

Data used from the IRCAD research center in France. This dataset has 20 annotated 3D CT series, out of which 15 used for training and the remaining 5 are used for testing.
Dataset link: https://www.dropbox.com/s/8h2avwtk8cfzl49/ircad-dataset.zip?dl=0<br>

### Download the data and unzip it and move all the .gz files inside "dataset/data"

```python
class NiiDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, image_filenames, mask_filenames, batch_size, image_size):
        self.image_filenames = image_filenames
        self.mask_filenames = mask_filenames
        self.batch_size = batch_size
        self.image_size = image_size

    def __len__(self):
        return int(np.ceil(len(self.image_filenames) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_x = self.image_filenames[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_y = self.mask_filenames[idx * self.batch_size:(idx + 1) * self.batch_size]

        x = np.zeros((self.batch_size, *self.image_size, 1))
        y = np.zeros((self.batch_size, *self.image_size, 1))

        for i, (image_filename, mask_filename) in enumerate(zip(batch_x, batch_y)):
            image = nib.load(image_filename)
            mask = nib.load(mask_filename)

            image_data = image.get_fdata()
            mask_data = mask.get_fdata()
            
            image_data = tf.image.resize(image_data, self.image_size)
            mask_data = tf.image.resize(mask_data, self.image_size)
            # get random slice from the volumes
            slice_index = random.randint(0, image_data.shape[2] - 1)
            x[i, :, :, 0] = image_data[:, :, slice_index]
            y[i, :, :, 0] = mask_data[:, :, slice_index]
        
        return x, y
```

```python
data_images = sorted(glob(os.path.join(data_path, 'data', '*_orig.nii.gz')))
data_masks = sorted(glob(os.path.join(data_path, 'data', '*_liver.nii.gz')))

batch_size = 5
image_size = (512, 512)

train_generator = NiiDataGenerator(data_images[:15], data_masks[:15], batch_size, image_size)
val_generator = NiiDataGenerator(data_images[15:], data_masks[15:], batch_size, image_size)
```

## Model

```python
def encoder(inputs, filters, pool_size):
    conv_pool = tf.keras.layers.Conv2D(filters, (3, 3), activation='relu', padding='same')(inputs)
    conv_pool = tf.keras.layers.MaxPooling2D(pool_size=pool_size)(conv_pool)
    return conv_pool

def decoder(inputs, concat_input, filters, transpose_size):
    up = tf.keras.layers.Concatenate()([tf.keras.layers.Conv2DTranspose(filters, transpose_size, strides=(2, 2), padding='same')(inputs), concat_input])
    up = tf.keras.layers.Conv2D(filters, (3, 3), activation='relu', padding='same')(up)
    return up

def UNet(img_size=(512,512,1)):
    inputs = tf.keras.Input(img_size)

    # Encoder
    conv_pool1 = encoder(inputs, 32, (2, 2))
    conv_pool2 = encoder(conv_pool1, 64, (2, 2))
    conv_pool3 = encoder(conv_pool2, 128, (2, 2))
    conv_pool4 = encoder(conv_pool3, 256, (2, 2))

    # Bottleneck
    bridge = tf.keras.layers.Conv2D(512, (3, 3), activation='relu', padding='same')(conv_pool4)

    #Decoder
    up6 = decoder(bridge, conv_pool3, 256, (2, 2))
    up7 = decoder(up6, conv_pool2, 128, (2, 2))
    up8 = decoder(up7, conv_pool1, 64, (2, 2))
    up9 = decoder(up8, inputs, 32, (2, 2))
    outputs = tf.keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(up9)
    model = tf.keras.Model(inputs=[inputs], outputs=[outputs])
    return model

model = UNet()
```

```python
def dice_coef(y_true, y_pred, smooth=1.):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    
    return (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)
```

```python
loss = sm.losses.JaccardLoss()

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=3e-4), loss=loss, metrics=[dice_coef])
```

```python
history = model.fit(train_generator,
                              steps_per_epoch=len(train_generator),
                              epochs=100,
                              validation_data=val_generator,
                              validation_steps=len(val_generator)
                   )
```

```python
#saving the model
save_model(model, "model/model.h5")
```

# Save Images and Mask

```python
import os
from glob import glob
from PIL import Image
```

```python
def get_images(img_paths, mask_paths):
    images = []
    masks = []
    for img_path, mask_path in zip(img_paths, mask_paths):
        image = nib.load(img_path)
        mask = nib.load(mask_path)
        
        image_data = image.get_fdata()
        mask_data = mask.get_fdata()
        
        for slice_idx in range(image_data.shape[-1]):
            img_slice = image_data[:,:,slice_idx]
            mask_slice = mask_data[:,:,slice_idx]
            
            images.append(img_slice)
            masks.append(mask_slice)
    
    return images, masks
```

```python
img_paths = sorted(glob(os.path.join(data_path, 'data', '*_orig.nii.gz')))
mask_paths = sorted(glob(os.path.join(data_path, 'data', '*_liver.nii.gz')))
```

```python
image_files, mask_files = get_images(img_paths, mask_paths)
```

```python
def save_data(image_files, mask_files):
    image_dir = "dataset/images"
    mask_dir = "dataset/masks"
    if not os.path.exists(image_dir):
        os.makedirs(image_dir)
        
    if not os.path.exists(mask_dir):
        os.makedirs(mask_dir)
        
    gray_cnvrtr = lambda x: x.convert('L')
    
    for idx, (image,mask) in enumerate(zip(image_files, mask_files)):
        pil_image = Image.fromarray(image)
        pil_mask = Image.fromarray(mask*255)
        
        pil_image = gray_cnvrtr(pil_image)
        pil_mask = gray_cnvrtr(pil_mask)
        
        pil_image.save(os.path.join(image_dir, f"img_{idx}.jpg"))
        pil_mask.save(os.path.join(mask_dir, f"mask_{idx}.jpg"))
```

```python
save_data(image_files, mask_files)
```
